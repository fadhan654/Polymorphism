```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Buat dataset sederhana
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Fitur
y = 4 + 3 * X + np.random.randn(100, 1)  # Target (dengan sedikit noise)
```

```python
# Plot dataset
plt.scatter(X, y)
plt.xlabel('Fitur')
plt.ylabel('Target')
plt.title('Dataset Sederhana untuk Latihan Regresi')
plt.show()
```

```python
# Inisialisasi model regresi linear
model = LinearRegression()
```

```python
# Latih model menggunakan data
model.fit(X, y)
```

```python
# Prediksi target
y_pred = model.predict(X)
```

```python
# Hitung R-squared
r2 = model.score(X, y)
```

```python
# Hitung RMSE, MAE, MSE, dan MAPE
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

rmse_score = rmse(y, y_pred)
mae_score = mae(y, y_pred)
mse_score = mse(y, y_pred)
mape_score = mape(y, y_pred)
```

```python
# Tampilkan hasil
print("Koefisien (slope):", model.coef_[0][0])
print("Intersep:", model.intercept_[0])
print("R-squared:", r2)
print("RMSE:", rmse_score)
print("MAE:", mae_score)
print("MSE:", mse_score)
print("MAPE:", mape_score)
```

```python
# Plot garis regresi
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Fitur')
plt.ylabel('Target')
plt.title('Regresi Linear dengan Garis Regresi')
plt.show()
```
